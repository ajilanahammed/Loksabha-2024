```python
import pandas as pd
import matplotlib.pyplot as plt

csv_file_path = '/kaggle/input/loksabha-2024/election_results_2024.csv'  
data = pd.read_csv(csv_file_path)


data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')

# Insight 1: Overall Winning Party Distribution
winning_party_distribution = data['Leading Party'].value_counts()
print("Winning Party Distribution:")
print(winning_party_distribution)
plt.figure(figsize=(12, 8))
winning_party_distribution.plot(kind='bar', color='red')
plt.title('Winning Party Distribution')
plt.xlabel('Party')
plt.ylabel('Number of Seats')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('winning_party_distribution.png')
plt.show()

# Insight 2: Top 5 Constituencies with Highest Victory Margin
top_5_highest_margins = data.nlargest(5, 'Margin')[['Constituency', 'Leading Candidate', 'Leading Party', 'Margin']]
print("\nTop 5 Constituencies with Highest Victory Margin:")
print(top_5_highest_margins)

plt.figure(figsize=(14, 10))
plt.bar(top_5_highest_margins['Constituency'], top_5_highest_margins['Margin'], color='black')
plt.title('Top 5 Constituencies with Highest Victory Margin')
plt.xlabel('Constituency')
plt.ylabel('Margin')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_5_highest_margins.png')
plt.show()

# Insight 3: Top 5 Constituencies with Lowest Victory Margin
top_5_lowest_margins = data.nsmallest(5, 'Margin')[['Constituency', 'Leading Candidate', 'Leading Party', 'Margin']]
print("\nTop 5 Constituencies with Lowest Victory Margin:")
print(top_5_lowest_margins)

plt.figure(figsize=(12, 8))
plt.bar(top_5_lowest_margins['Constituency'], top_5_lowest_margins['Margin'], color='red')
plt.title('Top 5 Constituencies with Lowest Victory Margin')
plt.xlabel('Constituency')
plt.ylabel('Margin')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_5_lowest_margins.png')
plt.show()

# Insight 4: Leading Candidates with the Largest Vote Margins
leading_candidates_largest_margins = data.nlargest(5, 'Margin')[['Leading Candidate', 'Leading Party', 'Margin']]
print("\nLeading Candidates with Largest Vote Margins:")
print(leading_candidates_largest_margins)

# Insight 5: Leading Candidates with the Smallest Vote Margins
leading_candidates_smallest_margins = data.nsmallest(5, 'Margin')[['Leading Candidate', 'Leading Party', 'Margin']]
print("\nLeading Candidates with Smallest Vote Margins:")
print(leading_candidates_smallest_margins)

# Insight 6: Distribution of Declared Results
results_status_distribution = data['Status'].value_counts()
print("\nDistribution of Declared Results:")
print(results_status_distribution)

plt.figure(figsize=(12, 8))
results_status_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Declared Results')
plt.ylabel('')
plt.tight_layout()
plt.savefig('results_status_distribution.png')
plt.show()

# Insight 7: Comparison of Major Parties (e.g., BJP vs INC)
major_parties_comparison = data[data['Leading Party'].isin(['Bharatiya Janata Party', 'Indian National Congress'])]['Leading Party'].value_counts()
print("\nComparison of Major Parties:")
print(major_parties_comparison)

plt.figure(figsize=(10, 6))
major_parties_comparison.plot(kind='bar', color=['orange', 'blue'])
plt.title('Comparison of Major Parties (BJP vs INC)')
plt.xlabel('Party')
plt.ylabel('Number of Seats')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('major_parties_comparison.png')
plt.show()

# Insight 8: Performance of New or Small Parties
small_parties_performance = data[data['Leading Party'].isin(small_parties)]['Leading Party'].value_counts()
print("Performance of New or Small Parties:")
print(small_parties_performance)


plt.figure(figsize=(10, 6))
small_parties_performance.plot(kind='bar', color='orange')
plt.title('Performance of New or Small Parties')
plt.xlabel('Party')
plt.ylabel('Number of Seats')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('small_parties_performance.png')
plt.show()


```
